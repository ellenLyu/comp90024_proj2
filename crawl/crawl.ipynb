```python
tweet_count = 1000
keyword = "plane fly OR car drive OR covid OR outbreak -filter:retweets"


def tweet_collector(geo, keyword, tweet_count):

    while tweet_count > 0:
        try:
            for page in tweepy.Cursor(api.search, q=keyword, include_rts=False, geocode=geo).pages():
                print("new page \n")
                for tweet in page:
                    #print(str(tweet._json))
                    print(tweet.text, "id:", tweet.id_str)
                    if tweet_count == 0:
                        print("Collection finished")
                        return None
                    tweet_count -= 1
        except TweepError as e:
            print(e)
            print(e.response)
            print(e.response.status_code)
    print("Collection finished")


tweet_collector(geo, keyword, tweet_count)
```
