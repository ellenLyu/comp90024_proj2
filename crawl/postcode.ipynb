```python
import pandas as pd
import json


file1 = "Australian_Post_Codes_Lat_Lon.csv"
file2 = "au-towns-sample.csv"

melb_geo_file = "melb_geo.json"
```

```python
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

with open(melb_geo_file, "r") as f:
    melb_geo = json.load(f)
```

```python
df1
```

```python
melb_geo
```

```python
melb_geo['features'][0]
```

```python
suburbs = [feature['properties']['name'] for feature in melb_geo['features']]
```

```python
suburbs
```

```python
df_vic = df1[df1['state'] == 'VIC']
```

```python
df_vic
```

```python
df_vic['suburb'] = df_vic['suburb'].apply(str.lower)
```

```python
df_vic
```

```python
temp = []
for suburb in suburbs:
    a = df_vic[df_vic['suburb'] == suburb.lower()]
    temp.append(a)
```

```python
df_melb = pd.concat(temp)
```

```python
df_melb
```

```python
suburb_coordinates = {}
for _, row in df_melb.iterrows():
    name = row['suburb']
    suburb_coordinates[name] = row['lat'], row['lon']
```

```python
suburb_coordinates
```

```python
with open("melb_coordinates.json", "w") as f:
    json.dump(suburb_coordinates, f)
```
